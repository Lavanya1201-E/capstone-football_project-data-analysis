To remove redunant columns (duplicate columns, empty columns, constant columns)

```python
df=pd.read_excel("Desktop/football project/final_merge_file.xlsx")
df
```

```python
#To find duplicate columns
duplicate_cols = df.columns[df.T.duplicated()]
print("Duplicate columns:", duplicate_cols.tolist())
```

df.T.duplicated() checks for rows that are identical after transposing (so columns become rows).

Columns with exactly the same values across all rows are considered duplicates.

```python
#To find empty columns
empty_cols = df.columns[df.isna().all()]
print("Empty columns:", empty_cols.tolist())
```

These columns are completely useless and can be dropped.

```python
#To find columns with constant value
constant_cols = [col for col in df.columns if df[col].nunique() <= 1]
print("Constant columns:", constant_cols)
```

df[col].nunique() counts unique values.

If it’s 1 (or 0), the column has the same value in all rows → redundant for analysis

```python
# drop all redundant columns at once
redundant_cols = list(duplicate_cols) + list(empty_cols) + list(constant_cols)
Football_data_merged_cleaned = df.drop(columns=redundant_cols)
```

To treat the null values

```python
import pandas as pd

# Assuming your dataframe is called df

# 1. Fill safe categorical metadata with 'Unknown'
safe_cats = [
    'home_club_manager_name',
    'away_club_manager_name',
    'referee',
    'home_club_name',
    'away_club_name'
]
for col in safe_cats:
    if col in df1.columns:
        df1[col] = df1[col].fillna('Unknown')

# 2. Fill attendance (numeric) with median
if 'attendance' in df1.columns:
    df1['attendance'] = df1['attendance'].fillna(df1['attendance'].median())

# 3. Fill positions with 'Unknown'
pos_cols = ['home_club_position', 'away_club_position']
for col in pos_cols:
    if col in df1.columns:
        df1[col] = df1[col].fillna('Unknown')

# 4. Leave event columns (game_event_id, date_y, minute, type_x,
# player_id_y, description, player_in_id, player_assist_id) as-is,
# because null means “no event”

# Optional: verify null counts again
print(df1.isna().sum())
```


 Interpretaion for null value treatment:
“I treated nulls differently based on their meaning.
For metadata like manager names, referee, club names, club positions and attendance, the missing values were not critical, so I filled them with ‘Unknown’ or median values to keep the dataset consistent.
But for event-level fields like game_event_id, minute, description or assist IDs, the nulls actually mean no event occurred, so I left them as null to preserve the real match information.”

Outlier treatment

```python

#  Select numeric columns
numeric_cols = df1.select_dtypes(include=['int64', 'float64']).columns

# 2️ Columns where extreme values are likely valid events
# We will skip removing/capping these
safe_event_cols = ['goals', 'assists', 'yellow_cards', 'red_cards', 'minutes_played']

#  Columns to check for outliers
check_cols = [col for col in numeric_cols if col not in safe_event_cols]

#  Outlier treatment using IQR capping
for col in check_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the outliers
    df1[col] = df1[col].clip(lower_bound, upper_bound)

#  check counts of outliers before and after
for col in check_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)]
    print(f"{col} has {outliers.shape[0]} outliers after capping")

# df1 is now cleaned of extreme outliers safely
```
